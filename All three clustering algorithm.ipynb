# Data Preparation

We have to normalize the variableto express them in the same range of values. In other words, normalization means adjusting values measured on different scales to a common scales.

```python
from sklearn.preprocessing import normalize 
scaled = normalize(df)
scaled = pd.DataFrame(scaled)
scaled.head()
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
sse = []
for i in range(1, 5):
    kmeans = KMeans(n_clusters = i, init = 'random', random_state = 42)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)
    
sse
```

```python
sse
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,5),sse)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('clustering score')
plt.show()
```

```python
#fit the model and predict
kmeans  = KMeans(n_clusters = 2, random_state = 42)
kmeans.fit(df)
```

```python
pred = kmeans.predict(df)
pred
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled, pred)
```

# Hierarichal Clustering

```python
import pandas as pd
```

```python
df = pd.read_csv("D:/Ranjeet Singh/Data Science/ML/winequality_red.csv")
df
```

# Data Preparation

```python
from sklearn.preprocessing import normalize 
scaled = normalize(df)
scaled = pd.DataFrame(scaled)
scaled.head()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
plt.figure(figsize = (10,7))

dendrogram = sch.dendrogram(sch.linkage(scaled, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Quality of wine')
plt.xlabel('apply all features')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
pred = cluster.fit_predict(scaled)
pred
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled, pred)
```

```python
import pandas as pd
df = pd.read_csv("D:/Ranjeet Singh/Data Science/ML/winequality_red.csv")
df.head()
```

```python
z = df.columns
z
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms
```

```python
df = ms.fit_transform(df)
df
```

```python
df=pd.DataFrame(df)
df
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans 
km = KMeans(n_clusters=2, random_state=0)
km.fit(df)
```

```python
sse = []
k_rng = range(1,5)
for k in k_rng:
    km = KMeans(n_clusters = k)
    km.fit(df)
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel('K')
plt.ylabel('sum of squared error')
plt.plot(k_rng,sse)
plt.show()
```

```python
km.cluster_centers_
```

```python
pred = km.predict(df)
pred
```

```python
km.inertia_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df,pred)
```
